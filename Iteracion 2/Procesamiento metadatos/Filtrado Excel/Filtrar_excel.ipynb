```python
import pandas as pd
import ast
```

```python
# Cargar el archivo Excel
df = pd.read_csv('/content/PADCHEST_chest_x_ray_images_labels_160K_01.02.19.csv')
```

```python
df.head(1)
```

```python
df['Labels']
```

```python
# Especifica el valor por el cual deseas filtrar en la columna "Labels"
valor_a_filtrar = 'pneumonia'

# Filtrar el DataFrame excluyendo NaN en la columna "Labels"
df_filtrado = df[df['Labels'].apply(lambda x: valor_a_filtrar.lower() in [label.lower() for label in ast.literal_eval(x)] if isinstance(x, str) else False)]


# Crear un nuevo archivo CSV con los datos filtrados
ruta_nuevo_csv = '/content/Padchest_neumonia.csv'
df_filtrado.to_csv(ruta_nuevo_csv, index=False)

print(f"Nuevo archivo creado en: {ruta_nuevo_csv}")
```
