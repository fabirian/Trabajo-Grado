```python
import re
import csv
import pandas as pd
```

```python
# Archivo de entrada
archivo_entrada = '/content/padchest_neumonia.zip.unzip-l.txt'

# Archivo de salida CSV
archivo_salida = 'resultados.csv'

# Expresión regular para encontrar el patrón deseado
patron = re.compile(r'\d+_\w+\.png')

# Lista para almacenar las referencias encontradas
referencias = []

# Leer el archivo de entrada
with open(archivo_entrada, 'r') as archivo:
    # Iterar sobre cada línea del archivo
    for linea in archivo:
        # Buscar coincidencias en la línea
        coincidencias = patron.findall(linea)

        # Agregar las coincidencias a la lista de referencias
        referencias.extend(coincidencias)

# Escribir las referencias en un archivo CSV
with open(archivo_salida, 'w', newline='') as csvfile:
    escritor_csv = csv.writer(csvfile)

    # Escribir encabezado
    escritor_csv.writerow(['id_imagen'])

    # Escribir referencias en el CSV
    for ref in referencias:
        escritor_csv.writerow([ref])
```

```python
# Contar la cantidad de referencias
cantidad_referencias = len(referencias)
print(f"Total de referencias encontradas: {cantidad_referencias}")
```

```python
# Archivo CSV con las referencias
archivo_referencias = 'resultados.csv'

# Archivo CSV para la búsqueda
archivo_busqueda = '/content/PADCHEST_chest_x_ray_images_labels_160K_01.02.19.csv'

# Archivo CSV de salida
archivo_salida = 'resultados_busqueda.csv'

# Leer las referencias desde el archivo CSV
with open(archivo_referencias, 'r') as csvfile:
    lector_csv = csv.reader(csvfile)
    referencias = [fila[0] for fila in lector_csv]

# Abrir el archivo CSV para búsqueda y escritura
with open(archivo_busqueda, 'r') as csvfile, open(archivo_salida, 'w', newline='') as csvfile_salida:
    lector_csv = csv.reader(csvfile)
    escritor_csv = csv.writer(csvfile_salida)

    # Escribir encabezados en el archivo de salida
    encabezados = next(lector_csv)
    escritor_csv.writerow(encabezados)

    # Buscar y escribir las filas que coinciden con las referencias
    for fila in lector_csv:
        if any(ref in fila for ref in referencias):
            escritor_csv.writerow(fila)
```
