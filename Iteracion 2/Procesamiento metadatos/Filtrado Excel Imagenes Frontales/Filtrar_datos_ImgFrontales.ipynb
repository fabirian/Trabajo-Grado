```python
import pandas as pd
```

```python
# Cargar el archivo CSV
file_path = '/content/resultados_busqueda.csv'
df = pd.read_csv(file_path)
```

```python
df['Projection']
```

```python
# Valor específico para filtrar columnas
valor_especifico = 'PA'

# Filtrar las filas que contienen el valor específico
df_filtrado = df.loc[df['Projection'] == valor_especifico]

# Guardar el nuevo DataFrame en un nuevo archivo CSV
nuevo_file_path = '/content/metadato_img_frontales.csv'
df_filtrado.to_csv(nuevo_file_path, index=False)

print(f"Archivo filtrado guardado en: {nuevo_file_path}")
```
