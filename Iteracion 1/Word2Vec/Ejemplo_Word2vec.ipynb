```python
from gensim.models import Word2Vec
from gensim.models import KeyedVectors
import re
```

```python
# Cargar el texto plano de ejemplo
with open("/content/Cuento.txt", "r", encoding="utf-8") as file:
    raw_text = file.read()

# Preprocesar el texto
def preprocess_text(text):
    # Eliminar caracteres especiales y números
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Convertir a minúsculas
    text = text.lower()
    # Dividir el texto en oraciones (puedes ajustar esto según tu texto)
    sentences = text.split('.')
    # Tokenizar cada oración
    tokenized_sentences = [sentence.split() for sentence in sentences if sentence]
    return tokenized_sentences

# Obtener oraciones tokenizadas
tokenized_text = preprocess_text(raw_text)

# Entrenar el modelo Word2Vec
model = Word2Vec(tokenized_text, vector_size=1000, window=10, min_count=1, workers=8)
```

```python
# Imprimir el diccionario de palabras (vocabulario)
print("Diccionario de palabras:")
for word in model.wv.index_to_key:
    print(word)
```

```python
import random

import random

def generate_coherent_text(model, seed_word, num_words_to_generate=100):
    generated_text = []

    for _ in range(num_words_to_generate):
        generated_text.append(seed_word)

        # Obtener las palabras similares basadas en la palabra actual
        similar_words = model.wv.most_similar(seed_word, topn=5)

        # Filtrar palabras similares para mejorar la coherencia
        filtered_similar_words = [word for word, _ in similar_words if word not in generated_text]

        # Elegir aleatoriamente la siguiente palabra entre las palabras similares filtradas
        if filtered_similar_words:
            next_word = random.choice(filtered_similar_words)
        else:
            # Si no hay palabras filtradas, elegir aleatoriamente entre todas las palabras similares
            next_word = random.choice(similar_words)[0]

        # Actualizar la palabra semilla para la siguiente iteración
        seed_word = next_word

    return generated_text
```

```python
generated_coherent_text = generate_coherent_text(model, seed_word="grifo", num_words_to_generate=100)
print(" ".join(generated_coherent_text))
```
